```python
from csv import reader

appleStore =  list(reader(open('AppleStore.csv')))
googleplayStore = list(reader(open('googleplaystore.csv')))

googlePlay_header =  googleplayStore[:1]
ios_header = appleStore[:1]
```

```python
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line between rows
        
    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))

```

```python
# Apple
print(ios_header)
print('\n')
explore_data(appleStore , 0, 3, True)
```

```python
# Android
print(googlePlay_header)
print('\n')
explore_data(googleplayStore, 0, 3, True)
```

Let's say we have a task to detect ditches in which the name of the application is repeated(name based)

Using the len method, you can easily count the number of elements in a given list.

But now we will try to delete duplicate entries.

```python
reviews_max = {}
for i in googleplayStore[1:]:
    name = i[0]
    n_reviews = float(i[3])
    if name in reviews_max and  reviews_max[name] < n_reviews:
         reviews_max[name] =  n_reviews
    elif  not   name in reviews_max:
         reviews_max[name] =  n_reviews
print(len(reviews_max ))
    
```

```python
android_clean = [] 
already_added = []
```

```python
for i in  googleplayStore[1:]:
    name = i[0]
    n_reviews = float(i[3])
    if n_reviews == reviews_max[name] and  name not  in  already_added:
        android_clean.append(i)
        already_added.append(name)

        
```

Removing Non-English Apps: Part One

```python
def only_eanglish(some_string):
    counter = 0
    for i in some_string:
        if ord(i) >  127:
            counter+=1
            if counter==3:
                return False
    return True
```

```python
only_eanglish('Instagram')
```

```python
only_eanglish('爱奇艺PPS -《欢乐颂2》电视剧热播')
```

```python
eanglish_android = []
for i in  googleplayStore[1:]:
    name = i[0]
    # only_eanglish(name)
    if only_eanglish(name):
        eanglish_android.append(i)
```
