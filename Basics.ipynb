Removing Non-English Apps: Part One

```python
def only_eanglish(some_string):
    counter = 0
    for i in some_string:
        if ord(i) >  127:
            counter+=1
            if counter==3:
                return False
    return True
```

```python
only_eanglish('Instagram')
```

```python
only_eanglish('爱奇艺PPS -《欢乐颂2》电视剧热播')
```

```python
eanglish_android = []
for i in  googleplayStore[1:]:
    name = i[0]
    # only_eanglish(name)
    if only_eanglish(name):
        eanglish_android.append(i)


eanglish_ios = []
for i in  appleStore[1:]:
    name = i[0]
    # only_eanglish(name)
    if only_eanglish(name):
        eanglish_ios.append(i)
```

Most Common Apps by Genre: Part Two

```python
def freq_table(dataset,index:int):
    freq_list = {}
    for i in  dataset:
        col_num = i[index]
        if col_num in freq_list:
            freq_list[col_num] += 1
        else:
            freq_list[col_num] = 1
    return  freq_list
    
```

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

Most Popular Apps by Genre on the App Store

```python
 freq_prime_genre_apple=  freq_table(free_apple_games,11)
```

```python
for  genre in  freq_prime_genre_apple:
    total = 0
    len_genre = 0
    for  app in free_apple_games:
        genre_app = app[11]
        if genre_app == genre:            
            n_ratings = float(app[6])
            total += n_ratings
            len_genre += 1
    avg_n_ratings = total / len_genre
    print(genre, ':', avg_n_ratings)

```
