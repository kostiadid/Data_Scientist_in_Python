Our aim in this project is to find mobile app profiles that are profitable for the App Store and Google Play markets. We're working as data analysts for a company that builds Android and iOS mobile apps, and our job is to enable our team of developers to make data-driven decisions with respect to the kind of apps they build.


```python
from csv import reader

appleStore =  list(reader(open('AppleStore.csv')))
googleplayStore = list(reader(open('googleplaystore.csv')))

googlePlay_header =  googleplayStore[:1]
ios_header = appleStore[:1]
```

```python
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line between rows
        
    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))

```

```python
# Apple
print(ios_header)
print('\n')
explore_data(appleStore , 0, 3, True)
```

```python
# Android
print(googlePlay_header)
print('\n')
explore_data(googleplayStore, 0, 3, True)
```
