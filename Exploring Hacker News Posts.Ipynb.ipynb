```python
ask_posts = []
show_posts = []
other_posts = []
```

```python
for post in hn:
    title = post[1]
    if title.lower().startswith("ask hn"):
        ask_posts.append(post)
    elif title.lower().startswith("show hn"):
        show_posts.append(post)
    else:
        other_posts.append(post)
        
print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))
    
```

```python
total_ask_comments = 0
for i in ask_posts:
    total_ask_comments+=(int(i[4]))

avg_ask_comments = total_ask_comments / len(ask_posts)
print(avg_ask_comments)

total_show_comments = 0
for i in show_posts:
    total_show_comments+=(int(i[4]))
avg_show_comments =  total_show_comments / len(show_posts)
print(avg_show_comments)
```

As we can see, posts with questions show more activity because they engage users to respond to the post.


```python
from datetime import datetime as dt

result_list = []
for post in ask_posts:
    result_list.append(
        [post[6], int(post[4])]
    )

    
counts_by_hour = {}
comments_by_hour = {}


for i in result_list:
    hour_str = (i[0])
    d = dt.strptime(hour_str,"%m/%d/%Y %H:%M")
    hour = d.hour
    
    if not hour in counts_by_hour: 
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = i[1]
    else:
        counts_by_hour[hour] += 1
        comments_by_hour[hour] += i[1]
    
        
    
```

```python
avg_by_hour = []

for hr in comments_by_hour:
    avg_by_hour.append([hr, comments_by_hour[hr] / counts_by_hour[hr]])

avg_by_hour
```

```python
swap_avg_by_hour = []
for i in  avg_by_hour:
    swap_avg_by_hour.append([i[1],i[0]])
swap_avg_by_hour
```

```python
sorted_swap = sorted(swap_avg_by_hour, reverse=True)
sorted_swap
```

```python
print('Top 5 Hours for Ask Posts Comments')
```

```python
better_hours = [0,0] 
for average, hour in sorted_swap[:5]:
    print(f"{hour}:00: {average:.2f} average comments per post")
    temp = [hour,average]
    
    if temp[1] > better_hours[1]:
        better_hours = temp
print('///////Based on our analysis, it is best to post in: ')    
print(better_hours)  
```
