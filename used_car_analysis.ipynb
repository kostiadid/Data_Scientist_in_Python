```python
autos['brand'].value_counts()
```

```python
top_20_brand_list = autos['brand'].value_counts().index[:20]
```

For further analysis, I selected the 20 most common brands from our dataset thanks to the value_counts() method.


```python
brand_mean_price_dict = {}
brand_mean_km_dict = {}
```

```python
for i in top_20_brand_list: 
    filteredbrand = autos[autos['brand']==i]
    avrg_price = filteredbrand['price'].mean()
    #print(avrg_price)
    brand_mean_price_dict[i] = int(avrg_price)
brand_mean_price_dict = dict(sorted(brand_mean_price_dict.items(),key=lambda x:x[1]))
brand_mean_price_dict 
```

Based on the analysis, we can say that although Volkswagen has the most models, it still has a fairly average price on the market, which makes it affordable for the mid-range class.

```python
bmp_series = pd.Series(brand_mean_price_dict )
print(bmp_series)
```

```python
mean_price_df = pd.DataFrame(bmp_series, columns=['mean_price'])
mean_price_df 
```

```python
for i in top_20_brand_list: 
    filteredbrand = autos[autos['brand']==i]
    avrg_km= filteredbrand['odometer_km'].mean()
    #print(avrg_price)
    brand_mean_km_dict[i] = int(avrg_km)
brand_mean_km_dict = dict(sorted(brand_mean_km_dict.items(),key=lambda x:x[1]))
brand_mean_km_dict 
```

```python
bm_km_series = pd.Series(brand_mean_km_dict )
mean_km_df = pd.DataFrame(bm_km_series, columns=['mean_km'])
```

```python
mean_km_df
```

Based on the mileage analysis, we see that although Volkswagen has an average price, its average mileage is higher than that of cheaper analogues, for example, Fiat.
